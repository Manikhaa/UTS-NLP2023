```python
import requests
from bs4 import BeautifulSoup
```

```python
import nltk
nltk.download('punkt')
import matplotlib.pyplot as plt
from nltk.tokenize import word_tokenize
from nltk.probability import FreqDist
```

```python
article_url = "https://setkab.go.id/pidato-presiden-ri-pada-penyampaian-keterangan-pemerintah-atas-ruu-tentang-apbn-tahun-anggaran-2023-beserta-nota-keuangannya-di-depan-rapat-paripurna-dpr-ri-di-gedung-nusantara-mpr-dpr-dpd-ri-senaya-2/"
```

```python
response = requests.get(article_url)
```

```python
if response.status_code == 200:
    # Menggunakan BeautifulSoup untuk mengekstrak teks
    soup = BeautifulSoup(response.text, 'html.parser')
    text = soup.get_text()
else:
    print("Gagal mengambil halaman web. Status respons:", response.status_code)
```

```python
words = word_tokenize(text)
```

```python
cleaned_words = [word.lower() for word in words if word.isalnum()]
```

```python
# Hitung frekuensi kata-kata
fdist = FreqDist(cleaned_words)
```

```python
top_words = fdist.most_common(10)
```

```python
plt.figure(figsize=(10, 5))
fdist.plot(10, cumulative=False)
plt.show()
```

```python
print("Kata kunci teratas:")
for word, freq in top_words:
    print(f"{word}: {freq}")
```
